```python
cat_train = X_train.select_dtypes(include = np.object)
cat_train.shape
```

```python
cat_train['coverage'] = cat_train['coverage'].map({'Basic' : 0, 'Extended' : 1, 'Premium' : 2})
cat_train['education'] = cat_train['education'].map({'Bachelor' : 1, 'College' : 1, 'High School or Below' : 0,
                                                     'Master': 2, 'Doctor':3})
cat_train['vehicle_size'] = cat_train['vehicle_size'].map({'Small' : 0, 'Medsize' : 1, 'Large' : 2})
cat_train['employmentstatus'] = cat_train['employmentstatus'].map({'Medical Leave': 0, 'Disabled' : 1,'Retired': 2,
                                                                   'Unemployed':3, 'Employed':4})
cat_train['location_code'] = cat_train['location_code'].map({'Urban' : 0, 'Suburban' : 1, 'Rural' : 2})



train_cat_encoded = cat_train[['coverage','education','vehicle_size','employmentstatus','location_code']]
```

```python
train_cat_onehot = X_train[['state','response','gender','marital_status', 'policy_type',
                            'renew_offer_type', 'sales_channel', 'vehicle_class']]
train_cat_onehot.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder().fit(train_cat_onehot)
train_onehot = encoder.transform(train_cat_onehot).toarray()
enc = [columname for sublist in encoder.categories_ for columname in sublist]
train_onehot = pd.DataFrame(train_onehot,columns=enc)
train_onehot.shape
train_onehot.head()
```
